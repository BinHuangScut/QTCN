```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

font = FontProperties()
font.set_family('serif')
font.set_name('Times New Roman')
font.set_size(12)

matrix1 = [[0.3291, 0.5158, 0.1551],
         [0.3252, 0.5323, 0.1424],
         [0.3346, 0.4873, 0.1782]]
matrix2 = [[0.3404, 0.4232, 0.2364],
         [0.3400, 0.4352, 0.2248],
         [0.3404, 0.4029, 0.2567]]
matrix3 = [[0.3214, 0.5465, 0.1321],
         [0.3157, 0.5652, 0.1190],
         [0.3296, 0.5138, 0.1566]]
matrix4 = [[0.3384, 0.4575, 0.2041],
         [0.3369, 0.4708, 0.1923],
         [0.3400, 0.4350, 0.2250]]

cmap = 'hot'

fig, axes = plt.subplots(2, 2, figsize=(8, 8))

data = [matrix1, matrix2, matrix3, matrix4]
min_val = np.min(data) 
max_val = np.max(data)

norm = Normalize(vmin=min_val, vmax=max_val)

sm = ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

im1 = axes[0, 0].imshow(matrix1, cmap=cmap, interpolation='nearest', norm=norm)
axes[0, 0].set_title('(a) Head 1',font = font, y= -0.15)
axes[0, 0].axis('off')

im2 = axes[0, 1].imshow(matrix2, cmap=cmap, interpolation='nearest', norm=norm)
axes[0, 1].set_title('(b) Head 2',font = font,y= -0.15)
axes[0, 1].axis('off')

im3 = axes[1, 0].imshow(matrix3, cmap=cmap, interpolation='nearest', norm=norm)
axes[1, 0].set_title('(c) Head 3',font = font,y= -0.15)
axes[1, 0].axis('off')

im4 = axes[1, 1].imshow(matrix4, cmap=cmap, interpolation='nearest', norm=norm)
axes[1, 1].set_title('(d) Head 4',font = font,y= -0.15)
axes[1, 1].axis('off')

cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
cbar = fig.colorbar(sm, cax=cbar_ax)

plt.savefig('Attention_Score.pdf', format='pdf',bbox_inches='tight')
plt.show()
```
